```python
import pandas as pd
data_files = {'Confirmed':'COVID/Confirmed.csv','Deaths':'COVID/Deaths.csv', 'Recovered':'COVID/Recovered.csv'}
```

```python
data_files = {'Confirmed':'COVID/Confirmed.csv','Deaths':'COVID/Deaths.csv', 'Recovered':'COVID/Recovered.csv'}
covid = {}
for key,value in data_files.items():
    covid[key] = pd.read_csv(value)
```

```python
confirmed = covid['Confirmed'].groupby('Country/Region').sum().copy()
confirmed = confirmed.reset_index()
death = covid['Deaths'].groupby('Country/Region').sum().copy()
death = death.reset_index()
recovered = covid['Recovered'].groupby('Country/Region').sum().copy()
recovered = recovered.reset_index()
```

```python
death = death.drop(['Lat','Long'],axis=1).T
death.columns = death.iloc[0,:]
death = death.drop(['Country/Region'],axis=0)

confirmed = confirmed.drop(['Lat','Long'],axis=1).T
confirmed.columns = confirmed.iloc[0,:]
confirmed = confirmed.drop(['Country/Region'],axis=0)

recovered = recovered.drop(['Lat','Long'],axis=1).T
recovered.columns = recovered.iloc[0,:]
recovered = recovered.drop(['Country/Region'],axis=0)
```
